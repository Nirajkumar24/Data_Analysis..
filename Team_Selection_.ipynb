```python
import pandas as pd

# Load the dataset
file_path = "/content/IPL"
data = pd.read_csv(file_path)

print("Dataset Overview:")
print(data.head())

print("\nDataset Summary:")
print(data.info())

print("\nDescriptive Statistics:")
print(data.describe())
```

```python
# Remove rows with missing values
data_cleaned = data.dropna()

# Remove duplicate rows
data_cleaned = data_cleaned.drop_duplicates()

#shape of the cleaned dataset
print("\nCleaned Dataset Shape:", data_cleaned.shape)
```

```python

# Batting Average
data_cleaned['batting_avg'] = data_cleaned['runs'] / data_cleaned['matches']

# Strike Rate
data_cleaned['batting_strike_rate'] = (data_cleaned['runs'] / data_cleaned['balls_faced']) * 100

# Total contribution
data_cleaned['Total Contribution'] = data_cleaned['runs'] + (data_cleaned['wickets'] * 20)

# Enhanced dataset
print("\nEnhanced Dataset with New Features:")
print(data_cleaned.head())
```

```python
# Sort players by Total Contribution
top_11_players = data_cleaned.sort_values(by='Total Contribution', ascending=False).head(11)

# Display the top 11 players
print("\nTop 11 Players:")
print(top_11_players[['player', 'Total Contribution']])  # Replace 'Player Name' with the correct column
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar chart for Total Contribution of top 11 players
plt.figure(figsize=(10, 6))
sns.barplot(
    x='player',
    y='Total Contribution',
    data=top_11_players
)
plt.title("Top 11 Players by Total Contribution")
plt.xticks(rotation=45)
plt.ylabel("Total Contribution")
plt.xlabel("Player Name")
plt.tight_layout()
plt.show()
```
